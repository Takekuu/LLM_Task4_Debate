#### 1. 数据处理

```python
from os import read
import pandas as pd
import numpy as np
import json

train_data_path = "./data/train_data.json"
test_data_path = "./data/test_data.json"
val_data_path = "./data/val_data.json"
train_data = json.load(open(train_data_path))
test_data = json.load(open(test_data_path))
val_data = json.load(open(val_data_path))


```

```python
def data_describe(frames):
    print(frames[0]['Topic'])
    for i in range(len(frames[0]['Debate'])):
        print(frames[0]['Debate'][i]['speaker'])
        print(frames[0]['Debate'][i]['content'])

print("train")
data_describe(train_data)
print("test")
data_describe(test_data)
print("val")
data_describe(val_data)
```

```python
import copy

def pre_processing_dataset(data):
    output = []
    input = copy.deepcopy(data)
    indexes = [-5, -3, -1]
    for debate in range(len(input)):
        temp = []
        for i in indexes:
            if abs(i) <= len(input[debate]['Debate']):
                temp.append(copy.deepcopy(input[debate]['Debate'][i]))
                input[debate]['Debate'][i]['speaker'] = ''
                input[debate]['Debate'][i]['content'] = ''
            else :
                print("Debate",debate,"\tTotal dabates count",len(input[debate]['Debate']),"\t Called debates index",i)
        output.append(temp)
    return input, output

                

input_corpus, output_corpus = pre_processing_dataset(train_data)
print(input_corpus[0])
print(output_corpus[0])
```

```python
instruction = "你是辩论专家，你能够根据指定格式的，包含辩论主题和辩论对话记录的辩论数据，推测出辩论数据中空缺的内容，并以json格式顺序输出空缺的内容"

df = pd.DataFrame({
    'input': input_corpus,
    'output': output_corpus,
    'instruction': instruction
})

df.to_csv('./data/train_lora.csv', index=False)

print("CSV file has been saved as 'output_file.csv'")
```
